Chargement des librairies et des données essentielles

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Exploration des données et affichages des 5 premières lignes de la dataset

```python
data = pd.read_csv("../data/kc_house_data.csv")
data.head()
```

Voir les informations sur la dataset

```python
data.info()
```

Voir les informations sur la dataset

```python
data.describe()
```

Visualisation de la relation surface / prix d'une maison

```python
sns.scatterplot(x=data["sqft_living"], y=data["price"])

plt.xlabel("Surface de la maison (sqft)")
plt.ylabel("Prix de la maison ($)")
plt.title("Relation entre la surface et le prix des maisons")

plt.show()
```

Choix des caractéristiques

```python
features = data[["bedrooms", "bathrooms", "sqft_living"]]

features.head()
```

Creation de la variable cible

```python
target = data["price"]

target.head()
```

Division des données (train/test)

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    features,
    target,
    test_size=0.2,
    random_state=42
)
```

Création du modèle

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

Entrainement du modèle

```python
model.fit(X_train, y_train)
```

Faire une prédiction

```python
import pandas as pd

house = pd.DataFrame(
    [[3, 2, 1500]],
    columns=["bedrooms", "bathrooms", "sqft_living"]
)

prediction = model.predict(house)

price = prediction[0]

print(f"Le prix est de {round(price):,} $")
```

Faire des prédictions sur le jeu de test

```python
predictions = model.predict(X_test)

predictions[:5]
```

Calcul de l'erreur du modèle MAE

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(y_test, predictions)

print(f"Erreur moyenne du modèle : ≈ {int(mae):,} $")
```

Graphique prix réel vs prix prédit

```python
plt.figure(figsize=(8,6))

plt.scatter(y_test, predictions)

plt.xlabel("Prix réel")
plt.ylabel("Prix prédit")
plt.title("Prix réel vs Prix prédit par le modèle")

plt.show()
```

Heatmap de la corrélation entre les variables

```python
corr = data.corr(numeric_only=True)

plt.figure(figsize=(12,8))
sns.heatmap(corr, cmap="coolwarm")

plt.title("Corrélation entre les variables")
plt.show()
```

Nouvelle sélection des features

```python
features = data[
[
"bedrooms",
"bathrooms",
"sqft_living",
"grade",
"sqft_above",
"view",
"waterfront"
]
]

target = data["price"]
```

Réentrainement des modèles avec les nouvelles features

```python
X_train, X_test, y_train, y_test = train_test_split(
    features,
    target,
    test_size=0.2,
    random_state=42
)

model = LinearRegression()
model.fit(X_train, y_train)
```

Import de random forest

```python
from sklearn.ensemble import RandomForestRegressor
```

Création d'un modèle random forest

```python
rf_model = RandomForestRegressor(
    n_estimators=500,
    random_state=42
)
```

Entrainement du modèle random forest

```python
rf_model.fit(X_train, y_train)
```

Faire des prédictions avec le modèle random forest et afficher les 5 premières prédictions

```python
rf_predictions = rf_model.predict(X_test)

rf_predictions[:5]
```

Calculer l'érreur du modèle MAE

```python
from sklearn.metrics import mean_absolute_error

rf_mae = mean_absolute_error(y_test, rf_predictions)

print(f"Erreur moyenne Random Forest : ≈ {int(rf_mae):,} $")
```

Graphique prix réel vs prix prédit avec le modèle random forest

```python
plt.figure(figsize=(8,6))

plt.scatter(y_test, rf_predictions)

plt.xlabel("Prix réel")
plt.ylabel("Prix prédit (Random Forest)")
plt.title("Prix réel vs Prix prédit - Random Forest")

plt.show()
```

Verification des variables qui influence le plus le modèle random forest

```python
import pandas as pd

importance = rf_model.feature_importances_

feature_importance = pd.DataFrame({
    "feature": features.columns,
    "importance": importance
})

feature_importance = feature_importance.sort_values(
    by="importance",
    ascending=False
)

feature_importance
```

```python
plt.figure(figsize=(8,5))

sns.barplot(
    data=feature_importance,
    x="importance",
    y="feature"
)

plt.title("Importance des variables dans le modèle")
plt.xlabel("Importance")
plt.ylabel("Variables")

plt.show()
```

Definition des la grille de paramètres pour le Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    "n_estimators": [100, 200],
    "max_depth": [10, 20, None],
    "min_samples_split": [2, 5],
    "min_samples_leaf": [1, 2]
}
```

Lancer la recherche

```python
grid_search = GridSearchCV(
    estimator=RandomForestRegressor(random_state=42),
    param_grid=param_grid,
    cv=3,
    n_jobs=-1,
    verbose=2
)

grid_search.fit(X_train, y_train)
```

Utiliser le meilleur modèle pour prédire les prix des maisons

```python
best_model = grid_search.best_estimator_
```

Evaluation du meilleur modèle

```python
best_predictions = best_model.predict(X_test)

best_mae = mean_absolute_error(y_test, best_predictions)

print(f"Erreur moyenne optimisée : ≈ {int(best_mae):,} $")
```
